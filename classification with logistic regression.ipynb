```
import pandas as pd
import numpy as np

train=pd.read_csv("train.csv")
plots=train["plots"].to_numpy()
labels=train["labels"].to_numpy()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df=0.2,max_features=10000, ngram_range=(1, 2))
features = tfidf.fit_transform(plots)
```

```
print(plots.shape)
print(features.shape)
```

```
from sklearn.linear_model import LogisticRegression

plots_features=(features.todense()).getA()
model= LogisticRegression(solver='lbfgs').fit(plots_features,labels)
print(model.score(plots_features,labels))
```

```
test=pd.read_csv("test.csv")
test_plots=test["plots"].to_numpy()
test_labels=test["labels"].to_numpy()
```

```
test_plots_features=tfidf.fit_transform(test_plots).todense().getA()
print(model.score(test_plots_features,test_labels))
```
