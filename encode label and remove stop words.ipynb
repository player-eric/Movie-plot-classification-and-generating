```
import pandas as pd

data=pd.read_csv("dataset.csv")
data=data[["Genre","Plot"]]
data.head()
```

```
plots=data["Plot"].to_numpy()
labels=data["Genre"].to_numpy()
```

```
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
labels=encoder.fit_transform(labels)
print(labels)
print(encoder.classes_)
```

```
import nltk
import re

nltk.download('stopwords')
from nltk.corpus import stopwords
print(len(set(stopwords.words('english'))))

plots_=[]
for plot in plots:
  plot=plot.lower()
  for stopword in set(stopwords.words('english')):
    plot=re.sub(" "+stopword+" "," ",plot)
    plot=re.sub("^"+stopword+" "," ",plot)
    plot=re.sub(stopword+"$"," ",plot)
  plots_.append(plot)

plots=plots_
print(plots[0])
```

```
import numpy as np
processed_data=pd.DataFrame({"plots":plots,"labels":labels})
```

```
processed_data.head()
```

```
from sklearn.model_selection import train_test_split

train, test = train_test_split(processed_data, test_size=0.2)
```
